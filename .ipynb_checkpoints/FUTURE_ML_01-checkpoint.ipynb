```python
# Step 0: config
DATA_PATH = 'data/raw/sales.csv'   # you said this file is in data/raw
FORECAST_PERIODS = 6               # months to forecast (change if you want 12)
AGG_FREQ = 'MS'                    # monthly start frequency; use 'D' for daily
HOLDOUT_PERIODS = FORECAST_PERIODS # last N periods held out for validation
EXPORT_FOLDER = 'exports'
```

```python
import os
print("CWD:", os.getcwd())
print("File exists:", os.path.exists(DATA_PATH))
print("Files in data/raw:", os.listdir('data/raw'))
# Show first 10 lines (safe preview)
with open(DATA_PATH, 'rb') as f:
    print(f.read(1000).decode('utf-8', errors='replace')[:1000])
```

```python
import pandas as pd

# Try latin1 (Windows encoding), fallback to errors='replace' if needed
try:
    df = pd.read_csv(DATA_PATH, encoding='latin1', low_memory=False)
except UnicodeDecodeError:
    df = pd.read_csv(DATA_PATH, encoding='ISO-8859-1', low_memory=False)

print("Initial shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

```python
# Normalize column names
df.columns = df.columns.str.strip().str.lower().str.replace(r'[^a-z0-9_]', '_', regex=True)
print("Cleaned columns:", df.columns.tolist())

# Quick peek
df.head()
```

```python
# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')

# Drop rows where date couldn't be parsed (just in case)
df = df.dropna(subset=['order_date'])

# Aggregate daily sales
daily_sales = df.groupby('order_date')['sales'].sum().reset_index()

print("Aggregated shape:", daily_sales.shape)
daily_sales.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(daily_sales['order_date'], daily_sales['sales'], color='blue', linewidth=1)
plt.title("Daily Sales Trend", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# Resample daily sales to monthly sales
monthly_sales = daily_sales.set_index('order_date').resample('ME').sum().reset_index()

print("Monthly sales shape:", monthly_sales.shape)
monthly_sales.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(monthly_sales['order_date'], monthly_sales['sales'], marker='o', color='green')
plt.title("Monthly Sales Trend", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
from prophet import Prophet

# Prepare data for Prophet: columns must be 'ds' (date) and 'y' (sales)
prophet_df = monthly_sales.rename(columns={'order_date':'ds', 'sales':'y'})

# Split into train/test if you want validation (optional)
FORECAST_PERIODS = 6  # months to forecast
train_df = prophet_df.copy()  # we can train on all data, or exclude last 6 months for validation

# Initialize and fit Prophet model
model = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
model.fit(train_df)

# Create future dataframe
future = model.make_future_dataframe(periods=FORECAST_PERIODS, freq='ME')  # 'M' is month-end
forecast = model.predict(future)

# Show last few forecasted values
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(FORECAST_PERIODS)
```
